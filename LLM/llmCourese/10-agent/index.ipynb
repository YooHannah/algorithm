### 4、「这」算不算 Agent?

**吴恩达**：“与其争论哪些工作才算是真正的 Agent，不如承认系统可以具有不同程度的 Agentic 特性。”

核心在于将复杂任务分解成多个步骤，并通过循环迭代的方式逐步优化结果。这种工作方式更接近于人类解决问题的思维模式：

- 目标设定: 明确任务目标；
- 规划分解: 将任务分解成多个子任务；
- 迭代执行: 依次执行每个子任务，并根据反馈结果进行调整和优化，最终完成目标。
